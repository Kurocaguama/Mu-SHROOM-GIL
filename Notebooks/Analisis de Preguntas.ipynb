```python
import pandas as pd
import random
```

```python
df_es = pd.read_json(r'C:\Users\madja\Desktop\Servicio Social\Mushroom\mushroom.es-train_nolabel.v1.jsonl', lines=True)
df_es
```

```python
# Da la pregunta y respuesta de la fila "index"
def info(index = int):
    columna = df_es.loc[index]
    return columna["model_input"], columna["model_output_text"], index
```

```python
# Esta función va a imprimir al azar una pregunta con respuesta e índice salvo que se especifique.
def sample(index = random.randint(0,491)):
    col = info(index)
    print(index)
    print("---")
    print(f"Pregunta: {col[0]}")
    print("---")
    print(f"Respuesta del modelo: {col[1]}")
```

```python
#Elige una pregunta al azar.
sample(random.randint(0,491))
```
